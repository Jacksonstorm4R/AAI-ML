# Q- Q Plot

```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
std_normal=np.random.normal(loc=0,scale=1,size=1000) #(mean, std-dev)
```

```python
std_normal
```

```python
# o to 100 th percentile of std-normal
for i in range(0,101):
    print(i,np.percentile(std_normal,i))
```

```python
# generate 100 samples from N(20,5)

measurements=np.random.normal(loc=20,scale=5,size=100)
```

```python
stats.probplot(measurements,dist="norm",plot=pylab)
pylab.show()
```

```python
#@ since bot lies on a line it is a gaussian 
```

```python
# lets change the sample sixe to be larger 
measurements=np.random.normal(loc=20,scale=5,size=100000)
stats.probplot(measurements,dist="norm",plot=pylab)
pylab.show()
```

```python
# lets change the sample sixe to be larger 
measurements=np.random.uniform(size=100000)
stats.probplot(measurements,dist="norm",plot=pylab)
pylab.show()
```
