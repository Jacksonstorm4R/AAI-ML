```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=pd.read_csv("iris.csv")
```

```python
iris_setos=iris[iris['species']=="setosa"]
iris_versicolor=iris[iris['species']=="versicolor"]
iris_virginica=iris[iris['species']=="virginica"]
```

```python
# 2D Density Plot, comtours plot
sns.jointplot(x="petal_length",y="petal_width",data=iris_setos,kind='kde')
plt.show()
```
